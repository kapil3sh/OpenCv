```python
cap = cv2.VideoCapture(0)

while True :
    ret, frame = cap.read()
    cv2.imshow('frame', frame)

    if cv2.waitKey(1) & 0xFF == ord('q') :
        break

cap.release()
#cv2.waitKey(0)
cv2.waitKey(5000) # 5 sec delay before image window closes
cv2.destroyWindow('frame')

    
```
